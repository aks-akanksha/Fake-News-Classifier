```python
import numpy as np
import pandas as pd
```

```python
fake = pd.read_csv("../input/fake-and-real-news-dataset/Fake.csv")
true = pd.read_csv("../input/fake-and-real-news-dataset/True.csv")
```

```python
fake.head()
```

```python
fake["fake"] = 1
```

```python
true.head()
```

```python
true["fake"] = 0
```

```python
frame = [fake, true]
data = pd.concat(frame)
```

```python
data.head()
```

```python
data.tail()
```

```python
data.drop(["date", "subject"], axis = 1, inplace = True)
```

```python
data.head()
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
input_arr = np.array(data["title"])
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

corpus = []

for i in range(0,40000):
    newArr = re.sub('[^a-zA-Z]', ' ', input_arr[i])
    
    newArr = newArr.lower()
    newArr = newArr.split()
    
    ps = PorterStemmer()
    
    newArr = [ps.stem(word) for word in newArr if not word in set(stopwords.words("english"))]
    
    newArr = " ".join(newArr)
    
    corpus.append(newArr)
```

```python
data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000)
x = cv.fit_transform(corpus).toarray()
y = data.iloc[0:40000, 2].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 7)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 1)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
